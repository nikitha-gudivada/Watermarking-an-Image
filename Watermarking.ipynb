```python
import cv2
from matplotlib import pyplot as plt
```

```python
img1=cv2.imread('w1.png')
```

```python
img2=cv2.imread('w2.jfif')
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
img1=cv2.resize(img1,(1000,1000))
```

```python
img2=cv2.resize(img2,(1000,1000))
```

```python
watermark=cv2.addWeighted(src1=img1,alpha=0.7,src2=img2,beta=0.2,gamma=0)
```

```python
plt.imshow(watermark)
```

```python
import cv2
image_file='w3.jfif'
image=cv2.imread(image_file)
overlay=image.copy()
output=image.copy()
height,width=image.shape[:2]
namestamp="Shot on  OnePlus"
cv2.putText(overlay,namestamp.format(0.3),(width-400,height-24),cv2.FONT_HERSHEY_DUPLEX,1.5,(255,255,255),thickness=2)
cv2.addWeighted(overlay,0.3,output,0.7,0,output)
cv2.imwrite("watermarkedtex.jpg",output)
```
